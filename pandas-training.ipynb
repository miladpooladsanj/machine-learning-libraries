# Problem Set 5
1. Plot the number of gold medals won by the US male and female athletes throughout the history of the Olympics.

```python
USA_medals = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')].groupby(['Edition', 'Gender'])
USA_medals.size().unstack('Gender', fill_value=0).plot();
```

# Data Visualizations
In this section, we want to present a summary of the total medals won by participating countries in the 2008 Olympics.

```python
Beijing_olympic = oo[oo.Edition == 2008]
g = Beijing_olympic.groupby(['NOC', 'Medal']).size().unstack('Medal', fill_value=0)
g = g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']]
```

```python
g = g.transpose()
plt.figure(figsize=(16,5))
sns.heatmap(g);
```

We next want to use plots to show the five athletes who have won the most gold medals over the history of the Olympics.

```python
g = oo.groupby(['Athlete', 'Medal']).size()
g = g.unstack('Medal', fill_value=0)
g = g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)
g = g[['Gold', 'Silver', 'Bronze']].head()
g.plot(kind='bar');
```

# Problem Set 6
1. In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline in the solution. 

```python
USA_medals = oo[oo.NOC == 'USA']
USA_g = USA_medals.groupby(['Edition', 'Athlete', 'Medal']).size()
USA_g = USA_g.unstack('Medal', fill_value=0)
USA_g['Total'] = USA_g['Gold'] + USA_g['Silver'] + USA_g['Bronze']
tu = []
for year, group in USA_g.groupby('Edition'):
    tu.append(group.sort_values('Total', ascending=False)[:1])
tu
```
